```python
import pandas as pd
import numpy as np
```

# Data Cleaning

```python
df = pd.read_csv("data/SAP Datasets.csv")
```

```python
df
```

```python
# Forward and backward filling
years_col = [str(i) for i in range(2000,2024)]
df[years_col] = df[years_col].apply(lambda row: row.ffill().bfill(), axis=1)
```

```python
df
```

```python
df_melted = df.melt(
    id_vars=['Country Code', "Indicator Code", "Topic"],
    value_vars=[str(i) for i in range(2000,2024)],
    var_name='Year',
    value_name='Value'
)

# Pivot the dataset
df_wide = df_melted.pivot(
    index='Country Code',                  # Each country becomes a row
    columns=['Indicator Code', 'Topic', 'Year'],  # These become new columns
    values='Value'                          # Fill the pivoted table with 'Value'
)

df_wide.columns = ['_'.join(map(str, col_tuple)) for col_tuple in df_wide.columns]
df_wide.reset_index(inplace=True)
```

```python
df_wide
```

```python
features = df_wide.drop('Country Code', axis=1).columns
df_wide[features] = df_wide[features].apply(lambda col: col.fillna(col.mean()), axis=0)
```

```python
df_wide
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (assuming df_wide is already loaded)
df = df_wide.copy()  # Make a copy to avoid modifying original

# Select numerical columns (excluding categorical ones like 'Country Code')
numeric_cols = df.drop('Country Code', axis=1).columns

# Step 2: Standardize the features (important for PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numeric_cols])

# Step 3: Apply PCA to retain 95% variance
pca = PCA(n_components=0.95)  # Retain 95% variance
principal_components = pca.fit_transform(scaled_data)

# Number of components selected
n_components = pca.n_components_
print(f"Number of Principal Components Retaining 95% Variability: {n_components}")

# Step 4: Determine Optimal Clusters using Elbow Method
wcss = []  # Within-Cluster Sum of Squares
K_range = range(1, 11)  # Testing K=1 to K=10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(principal_components)
    wcss.append(kmeans.inertia_)  # Inertia is the sum of squared distances to cluster center

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal k")
plt.show()


```

```python
principal_components.shape
```

```python
# Step 5: Apply K-Means Clustering on PCA-reduced data
optimal_k = 3  # Choose based on the Elbow method plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(principal_components)

# Step 6: Store Results
df['Cluster'] = clusters
```

```python
# Step 5: Apply K-Means Clustering on PCA-reduced data
optimal_k = 3  # Choose based on the Elbow method plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(principal_components)

# Step 6: Store Results
df['Cluster'] = clusters

# Display the clustered data
import ace_tools as tools
tools.display_dataframe_to_user(name="PCA & Clustering Results", dataframe=df)

# Step 7: Visualization (if PCA results are 2D or 3D)
if n_components == 2:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1], hue=clusters, palette='viridis')
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.title("Clusters based on PCA Components")
    plt.legend(title="Cluster")
    plt.show()
elif n_components == 4:
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2], c=clusters, cmap='viridis')
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.set_zlabel("PC3")
    plt.title("3D Cluster Visualization")
    plt.show()
```
