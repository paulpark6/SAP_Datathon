```python
import pandas as pd
```

```python
country_pre_rank_df = pd.read_csv('./data/country_initial_rankings.csv', index_col=0)
```

```python
indicator_result_df = pd.read_csv('./data/indicator_result_df.csv')
```

```python
country_pre_rank_df.head()
```

```python
# First, create a dictionary mapping Indicator Name -> Normalized Count
multiplier_map = dict(zip(indicator_result_df['Indicator Name'], 
                          indicator_result_df['Normalized Count']))

# Loop through each column in country_post_rank_df
for col in country_pre_rank_df.columns:
    if col in multiplier_map:
        country_pre_rank_df[col] *= multiplier_map[col]

country_post_rank_df = country_pre_rank_df.copy()
```

```python
country_post_rank_df['mean_rank'] = country_post_rank_df.drop('Country Name', axis=1).mean(axis=1)
```

```python
final_rankings = country_post_rank_df.sort_values('mean_rank', ascending=True)[['Country Name', 'mean_rank']].reset_index(drop=True)
```

```python
final_rankings['rank'] = final_rankings.index + 1
```

```python
final_rankings.to_csv('final_rankings.csv', index=False)
```
